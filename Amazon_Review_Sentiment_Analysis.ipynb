```python
def sentiment(label):
    if label == 5.0 or label == 4.0:
        return "Positive"
    elif label == 3.0:
        return "Neutral"
    elif label == 1.0 or label == 2.0:
        return "Negative"
```

```python
fd['overall'] = fd["overall"].apply(sentiment)
```

```python
fd.head()
```

```python
fd.overall.value_counts().plot(kind = "bar")
```

```python
fd["reviewText"] = fd["reviewText"].astype(str)
def toLower(text):
    return text.lower()

fd["reviewText"] = fd["reviewText"].apply(toLower)
```

```python
def data_preprocessing(text):
    text = re.sub(r"https\S+|www\S+https\S+","",text,flags = re.MULTILINE)
    text = re.sub(r"\@w+|\#","",text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
fd.reviewText = fd['reviewText'].apply(data_preprocessing)
```

```python
stemmer = PorterStemmer()
def data_preprocessing_stem(text):
    text = re.sub(r"https\S+|www\S+https\S+","",text,flags = re.MULTILINE)
    text = re.sub(r"\@w+|\#","",text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    stem_words = [stemmer.stem(w) for w in filtered_text]
    return " ".join(stem_words)
```

```python
fd.reviewText = fd['reviewText'].apply(data_preprocessing_stem)
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(fd['reviewText'])
```

```python
feature_names = vect.get_feature_names_out()
print('Number of features: {}\n'.format(len(feature_names)))
print('First 20 features: \n{}'.format(feature_names[:20]))
```

```python
X = fd['reviewText']
Y = fd['overall']
X = vect.transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 42)
```

```python
print('Size of x_train:',(x_train.shape))
print('Size of y_train:',(y_train.shape))
print('Size of x_test:',(x_test.shape))
print('Size of y_test:',(y_test.shape))
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred,y_test)
print('Test accuracy : {:.2f}%'.format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print('\n')
print(classification_report(y_test, logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels = logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = logreg.classes_)
disp.plot()
```

```python
param_grid = {'C':[0.001,0.01,1,10]}
grid = GridSearchCV(LogisticRegression(), param_grid)
grid.fit(x_train,y_train)
```

```python
print("Best parameters : ", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred,y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, y_pred))
print('\n')
print(classification_report(y_test,y_pred))
```
